```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
input_data=pd.read_excel('/content/drive/MyDrive/Entri - Data Science with Gen AI - Assignments/Capstone project 1 : Python capstone project/ABC Company.xlsx')
input_data
```

```python
input_data['Height']=np.random.randint(150,180,len(input_data))
print("Modified Inputdata sheet")
input_data
```

```python
# Preprocessing - Checking for NaN values in the table.

print(input_data.isnull().sum())
```

```python
input_data['College'] = input_data['College'].ffill()
input_data['Salary'] = input_data['Salary'].ffill()
```

```python
input_data
```

```python
print(input_data.isnull().sum())
```

```python
# 1. Distribution of the members in each team to be decided.
# print(input_data['Team'].unique())
plt.figure(figsize=(10, 10))
category_counts = input_data['Team'].value_counts()
unique_counts_df = category_counts.reset_index()
unique_counts_df.columns = ['Unique_Category', 'Count']
plt.pie(unique_counts_df['Count'], labels=unique_counts_df['Unique_Category'], autopct='%1.1f%%', startangle=90)
plt.title('Distribution of the team members in Team')
plt.show()
```

```python
# 2. Segregate employees based on their positions within the company.
# print(input_data['Position'].unique())
Count_of_employees=input_data.groupby('Position')['Name'].count()
print(Count_of_employees)
Salary_of_employees=input_data.groupby('Position')['Salary'].sum()
print(Salary_of_employees)
```

```python
# 3. Identify the predominant age group among employees.
input_data['Age'].hist(bins=7, color='skyblue', edgecolor='black')
plt.title("Histogram of Age")
plt.xlabel("Age range")
plt.ylabel("Number of players in this age range")
plt.show()
```

```python
# 4. Highest Salary: Team and position
salary_totals_Team = input_data.groupby(['Team'])['Salary'].sum().reset_index()
highest_spend_Team = salary_totals_Team.idxmax()
max_salary_Team = salary_totals_Team.max()
# print("Team with highest total salary expenditure:", highest_spend_Team)
# print("Total salary spent:", max_salary_Team)

salary_totals_Position = input_data.groupby(['Position'])['Salary'].sum().reset_index()
highest_spend_Position = salary_totals_Position.idxmax()
max_salary_Position = salary_totals_Position.max()
# print("Position with highest total salary expenditure:", highest_spend_Position)
# print("Total salary spent:", max_salary_Position)

salary_totals = input_data.groupby(['Team', 'Position'])['Salary'].sum().reset_index()
highest_spender = salary_totals.idxmax()      # gives (team, position)
max_salary = salary_totals.max()              # gives total amount
print("Team and position with highest salary expenditure:", highest_spender)
print("Total salary spent:", max_salary)
# Visualization using countplot
# print(salary_totals_Team)
# print(salary_totals_Position)
# print(salary_totals)

plt.figure(figsize=(15, 8))
plt.xlabel("Team", fontsize=3)
sns.barplot(x='Team', y='Salary', data=salary_totals, hue='Position')
plt.xticks(rotation=45)
plt.show()
```

```python
# 5.Correlation between age and salary
plt.figure(figsize=(10, 6))
input_data_Corr = input_data.select_dtypes(include="number")
correlation_matrix = input_data_Corr.corr()
sns.heatmap(correlation_matrix)
plt.title("Heatmap: Correlation Matrix")
plt.show()
```
